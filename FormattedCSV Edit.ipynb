```python
import pandas as pd
csv_file_path = r"D:\Qt\Finally\CONV\04-07-24.csv"
df = pd.read_csv(csv_file_path,header = None)
df
```

```python
start_date = input("Enter Start Date (dd-mm-yy) :- ")
start_time = input("Enter Start Time (hh:mm:ss) :- ")
```

```python
from datetime import datetime, timedelta

def check_date_increase(previous_date_str, current_date_str, date_format="%d-%m-%y"):
    previous_date = datetime.strptime(previous_date_str, date_format)
    current_date = datetime.strptime(current_date_str, date_format)
    
    # Check if the difference between the two dates is exactly one day
    return (current_date - previous_date).days == 1

# Convert the DataFrame column to a list of dates for easier processing
dates = df[0].tolist()
date_format = "%d-%m-%y"
# Initialize the new dates list with the start date
new_dates = [start_date]

# Iterate over the dates and adjust them according to the start date
for i in range(1, len(dates)):
    previous_new_date = new_dates[-1]
    if check_date_increase(dates[i-1], dates[i], date_format):
        # If the date has increased by one day, add one day to the previous new date
        new_date = (datetime.strptime(previous_new_date, date_format) + timedelta(days=1)).strftime(date_format)
    else:
        # If the date has not increased by one day, keep the new date the same
        new_date = previous_new_date
    
    new_dates.append(new_date)

# Update the DataFrame with the new dates
df[0] = new_dates
df
```

```python
def check_time_increase(previous_time_str, current_time_str, time_format="%H:%M:%S"):
    previous_time = datetime.strptime(previous_time_str, time_format)
    current_time = datetime.strptime(current_time_str, time_format)
    
    # Check if the current time is later than the previous time
    return current_time > previous_time

# Convert the DataFrame column to a list of times for easier processing
times = df[1].tolist()
time_format = "%H:%M:%S"
# Initialize the new times list with the start time
new_times = [start_time]

# Iterate over the times and adjust them according to the start time
for i in range(1, len(times)):
    previous_new_time = new_times[-1]
    if check_time_increase(times[i-1], times[i], time_format):
        previous_time = datetime.strptime(previous_new_time, time_format)
        interval = (datetime.strptime(times[i], time_format) - datetime.strptime(times[i-1], time_format))
        new_time = (previous_time + interval).strftime(time_format)
    else:
        # If the time has not increased, keep the new time the same
        new_time = previous_new_time
    
    new_times.append(new_time)

# Update the DataFrame with the new times
df[1] = new_times

# Print the updated DataFrame
df
```

```python
df.to_csv(csv_file_path, index=False, header=False)
print("Done")
```
