```python
import datetime
start_time_str = start_time1
mid_time_str = mid_time1
end_time_str = end_time1

start_time = datetime.datetime.strptime(start_time_str, '%H:%M:%S')
mid_time = datetime.datetime.strptime(mid_time_str, '%H:%M:%S')
end_time = datetime.datetime.strptime(end_time_str, '%H:%M:%S')

# Log-Time Set In Minutes
interval_minutes = 2

def generate_times(start_time, end_time, interval_minutes):
    times = []
    current_time = start_time
    while current_time <= end_time:
        random_seconds = random.randint(0, 5)
        adjusted_time = current_time + datetime.timedelta(seconds=random_seconds)
        times.append(adjusted_time.strftime('%H:%M:%S'))
        current_time += datetime.timedelta(minutes=interval_minutes)
    return times

fcol2 = generate_times(start_time, end_time, interval_minutes)

def find_nearest_time(target_time_str, times):
    target_time = datetime.datetime.strptime(target_time_str, '%H:%M:%S')
    nearest_time = min(times, key=lambda x: abs(datetime.datetime.strptime(x, '%H:%M:%S') - target_time))
    return nearest_time

nearest_time_to_mid = find_nearest_time(mid_time_str, fcol2)

length_when_mid_reached = fcol2.index(nearest_time_to_mid)

print(fcol2)
print(f"Nearest time to mid time ({mid_time_str}): {nearest_time_to_mid}")
print(f"Index when mid time is reached: {length_when_mid_reached}")

if length_when_mid_reached > 0:
    print("Previous time:", fcol2[length_when_mid_reached - 1])
if length_when_mid_reached < len(fcol2) - 1:
    print("Next time:", fcol2[length_when_mid_reached + 1])

len1 = length_when_mid_reached
len1
```

```python
import numpy as np
import pandas as pd
import random

# Function to generate random values in increasing or decreasing order
def generate_random_values(start, end, num, increasing=True):
    values = [random.uniform(start, end) for _ in range(num)]
    if increasing:
        values.sort()
    else:
        values.sort(reverse=True)
    return np.round(values, 1)

# Function to create combined DataFrame for a single column set
def create_combined_df(start_value, mid_value, end_value, len1, len2):
    # Generate random increasing values from start_value to mid_value
    readings1 = generate_random_values(start_value, mid_value, len1)
    
    # Generate random decreasing values from mid_value to end_value
    readings2 = generate_random_values(mid_value, end_value, len2, increasing=False)
    
    # Create DataFrames for each part of readings
    df1 = pd.DataFrame({'Readings': readings1})
    df2 = pd.DataFrame({'Readings': readings2})
    
    # Concatenate all DataFrames together
    combined_df = pd.concat([
        pd.DataFrame({'Readings': [start_value]}), 
        df1, 
        pd.DataFrame({'Readings': [mid_value]}), 
        df2, 
        pd.DataFrame({'Readings': [end_value]})
    ], ignore_index=True)
    
    return combined_df

# Assuming col3 to col8, acol3 to acol8, and ecol3 to ecol8 are defined as DataFrames with required values
start_values = [
    float(col3.iloc[0, 0]), float(col4.iloc[0, 0]), float(col5.iloc[0, 0]), 
    float(col6.iloc[0, 0]), float(col7.iloc[0, 0]), float(col8.iloc[0, 0])
]
mid_values = [
    float(acol3.iloc[0, 0]), float(acol4.iloc[0, 0]), float(acol5.iloc[0, 0]), 
    float(acol6.iloc[0, 0]), float(acol7.iloc[0, 0]), float(acol8.iloc[0, 0])
]
end_values = [
    float(ecol3.iloc[0, 0]), float(ecol4.iloc[0, 0]), float(ecol5.iloc[0, 0]), 
    float(ecol6.iloc[0, 0]), float(ecol7.iloc[0, 0]), float(ecol8.iloc[0, 0])
]

# Define the number of readings for each segment
len1 = n  # Example value, replace with your desired number
len2 = no - len1  # Example value, replace with your desired number

# Create combined DataFrames for each column set
combined_dfs = []
for i in range(6):
    combined_dfs.append(create_combined_df(start_values[i], mid_values[i], end_values[i], len1, len2))

# Special handling for col9
start_value_col9 = float(col9.iloc[0, 0]) 
mid_value_col9 = float(acol9.iloc[0, 0])  # Desired mid value for col9
end_value_col9 = float(ecol9.iloc[0, 0])  # Example end value, adjust as necessary

# Create combined DataFrame for col9
len1_col9 = len1
len2_col9 = len2

# Generate random increasing values from start_value_col9 to mid_value_col9
readings1_col9 = generate_random_values(start_value_col9, mid_value_col9, len1_col9)
# Generate random decreasing values from mid_value_col9 to end_value_col9
readings2_col9 = generate_random_values(mid_value_col9, end_value_col9, len2_col9, increasing=False)

# Create DataFrames for each part of readings for col9
df1_col9 = pd.DataFrame({'Readings': readings1_col9})
df2_col9 = pd.DataFrame({'Readings': readings2_col9})

# Concatenate all DataFrames together for col9
df_col9 = pd.concat([
    pd.DataFrame({'Readings': [start_value_col9]}), 
    df1_col9, 
    pd.DataFrame({'Readings': [mid_value_col9]}), 
    df2_col9, 
    pd.DataFrame({'Readings': [end_value_col9]})
], ignore_index=True)

# Add the DataFrame for col9 to the combined DataFrames
combined_dfs.append(df_col9)

# Generate col10 with 4-5 degree gap behind col9
def generate_col10_values(col9_values):
    col10_values = []
    for val in col9_values:
        gap = random.uniform(1, 2)
        col10_values.append(round(val - gap, 1))
    return col10_values

# Create col10 DataFrame
col9_values = df_col9['Readings'].values
col10_values = generate_col10_values(col9_values)
df_col10 = pd.DataFrame({'Readings': col10_values})

# Add the DataFrame for col10 to the combined DataFrames
combined_dfs.append(df_col10)

# Concatenate all combined DataFrames along the columns axis
final_df = pd.concat(combined_dfs, axis=1)

# Set display options and show the final combined DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
final_df
```
